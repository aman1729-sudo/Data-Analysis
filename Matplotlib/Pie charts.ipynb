```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
arr = [60,40,30,29]
labels = ["sixty","fourty","E1","E2"]
colors = ['blue','red','yellow','green']
plt.pie(arr,labels = labels,colors = colors,wedgeprops = {'edgecolor':'yellow'})
plt.title("Pie charts")
```
