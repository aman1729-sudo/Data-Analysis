```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

Import Data

```python
cars_df = sns.load_dataset("car_crashes")
cars_df
```

Distribution Plots

```python
sns.distplot(cars_df["not_distracted"]) #way to look at univariate distribution
#line represent kde = kernel density estimation (value at points , density of data) 
```

```python
sns.distplot(cars_df["not_distracted"],kde = False,bins = 25) #bins divide data
```

Joint Plots

```python
#used to compare 2 distribution and scatter plot by default
sns.jointplot(x = "speeding",y = "alcohol",data = cars_df,kind = "reg")
```

```python
sns.jointplot(x = "speeding",y = "alcohol",data = cars_df,kind = "kde")
```

```python
sns.jointplot(x = "speeding",y = "alcohol",data = cars_df,kind = "hex")
```

kde plots

```python
sns.kdeplot(cars_df["alcohol"])
```

Pair plots

```python
#plot relation across whole dataframe in numerical values
sns.pairplot(cars_df)
```

```python
tips_df = sns.load_dataset("tips")
sns.pairplot(tips_df,hue="sex",palette = "Blues")#with hue we can pass categorical data
```

Rug plot

```python
sns.rugplot(tips_df["tip"]) #give dataframe as sticks and more dense more values
```

Styling

```python
sns.set_style("darkgrid") #dark darkgrid white whitegrid ticks
plt.figure(figsize=(8,4))
sns.set_context("paper",font_scale= 1.8) #poster paper talk
sns.jointplot(x = "speeding",y = "alcohol",data = cars_df,kind="reg")
sns.despine(left = True, right = True) #removing both axis
```

Bar Plots

```python
sns.barplot(x = "sex",y = "total_bill",data= tips_df,estimator=np.var) #np.var np.cov np.med all used to estimate data 
```

Count Plots

```python
sns.countplot(x = "sex",data = tips_df) #gives total number of males and females
```

Bxx Plots

```python
#allowwus to compare different variables
sns.boxplot(x = "day",y = "total_bill",data = tips_df,hue = "sex")
plt.legend(loc = 1)
```

Violin Plot

```python
sns.violinplot(x= "day",y = "total_bill",data = tips_df,hue = "sex") #uses lde estimation of data points
```

```python
sns.violinplot(x= "day",y = "total_bill",data = tips_df,hue = "sex",split = True)
```

Strip True

```python
plt.figure(figsize=(8,5))

sns.stripplot(x = "day",y = "total_bill",data = tips_df,jitter=True,hue = "sex",dodge = True)
#stripplot shows data by refrencing one variable , jitter scatter the points , hue distributes data according to provided attribute and dodge completely separetes the data
```

Swarn Plots

```python
#combine capablities of many plots together
sns.violinplot(x = "day",y = "total_bill",data = tips_df)
sns.swarmplot(x = "day",y = "total_bill",data = tips_df,color ="white")
```

```python
sns.swarmplot(x = "day",y = "total_bill",data = tips_df)
```

Pallets

```python
plt.figure(figsize=(8,7))
#palette provides color to ur graph
sns.set_style("dark")
sns.set_context("talk")
sns.stripplot(x = "day", y = "total_bill",data = tips_df, hue = "sex",palette=  "magma")
```

Heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.4)
cars_dd = cars_df.corr() #corr tells how influenteial a variable on results
sns.heatmap(cars_dd,annot = True, cmap = "Blues")
```

```python
flight = sns.load_dataset("flights")
flight = flight.pivot_table(index = "month",columns = "year",values = "passengers")
sns.heatmap(flight , cmap = "Blues",linecolor= "white",linewidths=1)
```

Cluster Maps

```python
iris = sns.load_dataset("iris")
species = iris.pop("species")
sns.clustermap(iris)
```

```python
sns.clustermap(flight,cmap = "Blues",standard_scale = 1)
```

Pair Grid

```python
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris , hue="species")
iris_g.map(plt.scatter)
```

Facet Grid

```python
#print multiple plot in grid
tips_fg = sns.FacetGrid(tips_df,col = "time",row = "smoker")
tips_fg.map(plt.hist,"total_bill",bins = 8)
```

```python
tips_fg = sns.FacetGrid(tips_df,col = "time",row = "smoker")
tips_fg.map(plt.scatter,"total_bill","tip")
```

```python
att = sns.load_dataset("attention")
att_fg = sns.FacetGrid(att,col = "subject",col_wrap=5,height=1.5)
att_fg.map(plt.plot,"solutions","score",marker = ".")
```

regression Plots

```python
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.4)
sns.lmplot(x = "total_bill",y = "tip",hue="sex",data = tips_df,markers=['o','^'],
          scatter_kws = {'s':100,'linewidth':0.5,'edgecolor':'w'})
```
